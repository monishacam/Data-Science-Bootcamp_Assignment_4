### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#total number of players
total_players = len(purchase_data['SN'].value_counts())
total_frame_df = pd.DataFrame({"Total Players":[total_players]})
total_frame_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique items
unique_items = len(purchase_data['Item ID'].unique())

#Average price
average_price = purchase_data['Price'].mean()

#total number of purchases
total_purchases = len(purchase_data['Purchase ID'].value_counts())

#total revenue
total_revenue = purchase_data['Price'].sum()

#data frame to display data
frame_df = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": [average_price],
    "Number of Purchases": [total_purchases],
    "Total Revenue": [total_revenue]
})

#convert price columns to dollar format
frame_df["Average Price"] = frame_df["Average Price"].map("${:.2f}".format)
frame_df["Total Revenue"] = frame_df["Total Revenue"].map("${:.2f}".format)

#display frame
frame_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#total number of players to use to calculate percentages later
total_players = len(purchase_data['SN'].value_counts())

#count of male players
males = purchase_data.groupby('Gender').get_group('Male')
male_count = len(males['SN'].unique())
#percentage of male players
percent_males = round((male_count/total_players)*100, 2)

#count of female players
females = purchase_data.groupby('Gender').get_group('Female')
female_count = len(females['SN'].unique())
#percentage of female players
percent_females = round((female_count/total_players)*100, 2)

#count of others/non-disclosed players
others = purchase_data.groupby('Gender').get_group('Other / Non-Disclosed')
others_count = len(others['SN'].unique())
#percentage of other players
percent_others = round((others_count/total_players)*100, 2)

frame_2_df = pd.DataFrame({
    'Gender':['Male', 'Female', 'Other / Non-Disclosed'],
    'Total Count': [male_count, female_count, others_count],
    'Percent of Players': [percent_males, percent_females, percent_others]
})

frame_2_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count by gender
males_purchase_count = len(males)
females_purchase_count = len(females)
others_purchase_count = len(others)

#average purchase price by gender
average_male = round(males["Price"].sum()/males_purchase_count,2)
average_female = round(females["Price"].sum()/females_purchase_count,2)
average_others = round(others["Price"].sum()/others_purchase_count,2)

#total purchase value
total_male = round(males["Price"].sum(),2)
total_female = round(females["Price"].sum(),2)
total_others = round(others["Price"].sum(),2)

#Average Purchase Total per Person
per_male = round(total_male/males_purchase_count,2)
per_female = round(total_female/females_purchase_count,2)
per_others = round(total_others/others_purchase_count,2)

#dataframe to display data
frame_3_df = pd.DataFrame({
    'Gender':['Male','Female','Other / Non-Disclosed'],
    'Purchase Count':[males_purchase_count, females_purchase_count, others_purchase_count],
    'Average Purchase Price': [average_male, average_female, average_others],
    'Total Purchase Value': [total_male, total_female, total_others],
    'Avg Total Purchase per Person': [per_male, per_female, per_others]
})

#format columns of dataframe
frame_3_df["Average Purchase Price"] = frame_3_df["Average Purchase Price"].map("${:.2f}".format)
frame_3_df["Total Purchase Value"] = frame_3_df["Total Purchase Value"].map("${:.2f}".format)
frame_3_df["Avg Total Purchase per Person"] = frame_3_df["Avg Total Purchase per Person"].map("${:.2f}".format)

frame_3_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#get info
sn_purchase_count = purchase_data.groupby("SN").count()["Price"].rename("Purchase Count")
sn_avg_purchase_price = purchase_data.groupby("SN").mean()["Price"].rename("Average Purchase Price")
sn_total_purchase_value = purchase_data.groupby("SN").sum()["Price"].rename("Total Purchase Value")

# Convert to DataFrame
sn_purchased = pd.DataFrame({"Purchase Count": sn_purchase_count,
                                   "Average Purchase Price": round(sn_avg_purchase_price,2),
                                   "Total Purchase Value": sn_total_purchase_value})

top_spenders = sn_purchased.sort_values("Total Purchase Value", ascending=False)
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#get info
items_purchase_count = purchase_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_price = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
items_total_value = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

# Convert to DataFrame
items_purchased = pd.DataFrame({"Purchase Count":items_purchase_count,
                                   "Item Price": items_price,
                                   "Total Purchase Value":items_total_value,})

most_popular_items = items_purchased.sort_values("Purchase Count", ascending=False)
most_popular_items.head()
```
